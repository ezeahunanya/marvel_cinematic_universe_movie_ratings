- Score type column have many missing values so seem useless

- Audience type column is named incorrectly

Fix 8: Audience type variables needs cleaning 

```python
df["audience_type"].replace({"tomatometerAllCritics": "All Critics", 
                             "tomatometerTopCritics": "Top Critics", 
                             "audienceAll": "All Audience"}, inplace=True)
```

Test

Fix 9&10:

- Some of the data types are incorrect for example the score column is as an object which is a string but should be an integer
- Box office and runtime columns need to be converted into a usable format

```python
df.box_office_gross_usa = (df[~df.box_office_gross_usa.isnull()]
                           .box_office_gross_usa.apply(lambda x: x[1:-1]))
```

```python
df.box_office_gross_usa.head()
```

```python
df[["runtime_hour", "runtime_min"]] = df.runtime.str.extract(r'(\d+)h (\d+)m')
```

```python
df.drop(columns='runtime', inplace=True)
```

```python
df.head()
```

```python
df.release_date_theaters = df.release_date_theaters.astype('str')

df = df.astype(dtype={'movie_title': 'category', 'audience_type': 'category',
                 'score': 'int64', 'average_rating': 'float', 'score_sentiment':
                 'category', 'review_count': 'int64', 'rating_count':
                 'int64', 'liked_count': 'int64', 'not_liked_count':
                 'int64', 'release_date_theaters': 'datetime64[ns]', 
                 'box_office_gross_usa': 'float', 'runtime_hour': 'int64', 
                 'runtime_min': 'int64', 'tomatometer_state': 'category', 
                 'audience_class': 'category'}, errors='ignore')
```

```python
df.info()
```

```python
df['runtime_minutes'] = (df.runtime_hour * 60) + df.runtime_min
```

```python
df.drop(columns = ['runtime_hour', 'runtime_min'], inplace = True)
```

Reshaping dataframe; seperating scores by audience types.

```python
df_all_critics = df[df.audience_type == 'All Critics'].reset_index(drop=True).drop(columns 
        = ['audience_type', 'audience_class'])
```

```python
df_top_critics = df[df.audience_type == 'Top Critics'].reset_index(drop=True).drop(columns 
        = ['audience_type', 'audience_class'])
```

```python
df_all_audience = df[df.audience_type == 'All Audience'].reset_index(drop=True).drop(columns 
        = ['audience_type', 'tomatometer_state'])
```

```python
df1 = pd.merge(df_all_critics, df_top_critics, how='inner', on=(['movie_title', 
             'release_date_theaters', 'box_office_gross_usa', 'runtime_minutes']), suffixes = ('_all_critics', '_top_critics'))

df1 = pd.merge(df1, df_all_audience, how='inner', on=(['movie_title', 
             'release_date_theaters', 'box_office_gross_usa', 'runtime_minutes']))

df1.head()
```

```python
df1.columns
```

```python
df1.rename(columns={"score": "score_all_audience", "average_rating": "average_rating_all_audience",
                   "score_sentiment": "score_sentiment_all_audience", "review_count":
                   "review_count_all_audience", "rating_count": "rating_count_all_audience",
                   "liked_count": "review_count_all_audience", "not_liked_count":
                   "not_liked_count_all_audience", "audience_class": "audience_class"}, inplace=True)
```

```python
df1['movie_rank_all_critics'] = df1['average_rating_all_critics'].rank(method='first', ascending=False)
df1['movie_rank_top_critics'] = df1['average_rating_top_critics'].rank(method='first', ascending=False)
df1['movie_rank_all_audience'] = df1['average_rating_all_audience'].rank(method='first', ascending=False)
```

```python
df1 = df1.astype(dtype={'movie_rank_all_critics': 'int64', 'movie_rank_top_critics': 
                       'int64', 'movie_rank_all_audience': 'int64'})
```

## Conclusions

There is quite a large agreement between the critics and the audience, with 96% of MCU films being rated positively by at least 60% of both groups. There is only one exception, Captain marvel, which although is rated positively by 79% of critics it only has a positive rating from 47% of the audience. 40% of movies are above the median for both the crtics and the audience. There are differences in average ratings for each movie between the groups. 'Spider-Man: Far From Home' is ranked the highest for the audience with a score of 4.53/5 while its Black Panther for the critics with a score of 8.30/10. However, 'Avengers: Endgame' is rated second higest by both groups. For both groups, there is a clear positive relationship between the average ratings and the box office earnings.
