```python
def aggregate_reviews(df):
    '''Aggregate reviews by movie and sentiment into single rows'''
        
    dic = {}
    movies = []
    review_sentiments = []
    reviews = []

    for movie in df.movie_title.unique():
        for sentiment in df.review_sentiment.unique():
            movies.append(movie)
            review_sentiments.append(sentiment)
            reviews.append(list(df[(df.movie_title == movie) & (df.review_sentiment == sentiment)].review))

    dic['movie_title'] = movies
    dic['review_sentiment'] = review_sentiments
    dic['reviews'] = reviews
    
    df1 = pd.DataFrame.from_dict(dic)
    
    return df1
```

```python
df1 = aggregate_reviews(df)
df1.head()
```

```python
df1.set_index(['movie_title', 'review_sentiment'], inplace=True)
```

```python
df1.to_pickle("corpus.pkl")
```

```python
import re
import string

def clean_text(text):
    '''Remove punctuation, make text lowercase and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
df1.reviews = df1.reviews.astype(str)
df1.reviews = df1.reviews.apply(lambda x: clean_text(x))
```

```python
df1.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk import word_tokenize          
from nltk.stem import WordNetLemmatizer 
class LemmaTokenizer:
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, doc):
        return [self.wnl.lemmatize(t) for t in word_tokenize(doc)]

cv = CountVectorizer(tokenizer=LemmaTokenizer(), stop_words='english')
```

```python
reviews_cv = cv.fit_transform(df1.reviews)
reviews_dtm = pd.DataFrame(reviews_cv.toarray(), columns=cv.get_feature_names())
reviews_dtm.index = df1.index
reviews_dtm
```
